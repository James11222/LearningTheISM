# Acquire Data

We pull data from the CATS collaboration which hosts MHD ISM simulations with varying magnetic field strength and pressure values corresponding to different Mach Numbers and Alfv√©nic Mach Numbers. 

We will use this data in our analysis to study the ISM with the 4PCF.

$$\mathcal{M}_s \sim 0.7$$
$$\mathcal{M}_A \sim 0.7$$

```python
# !wget https://users.flatironinstitute.org/~bburkhart/data/CATS/MHD/256/b1p1/t_800/dens_t800.fits.gz
# !ls

for i in range(500, 901, 50):
    os.system(f'wget https://users.flatironinstitute.org/~bburkhart/data/CATS/MHD/256/b1p1/t_{i}/dens_t{i}.fits.gz')
    
os.system(f'mv dens_*.fits.gz Ms_0p7__Ma_0p7')
```
